### 1부터 입력받은 숫자까지 더하는 프로그램 (함수 사용)

```python
# 함수 정의
def sum_pro(b) :
    c=[]
    for i in range(1,b+1):
        c.append(i)
        result=sum(c)
    return result
# input 함수 사용과 숫자 입력받기
a = input('숫자를 하나 입력하세요 ')
#입력받은 숫자 정수로 변환
b = int(a)
#입력한 숫자까지의 합출력 
print("입력한 숫자까지의 합은:",sum_pro(b),"입니다")
```
